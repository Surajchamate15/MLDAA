```python
df.dropna(inplace =True)
```

```python
df.drop(['Email No.'],axis=1,inplace=True)
```

```python
df.head()
```

```python
x = df.drop(['Prediction'],axis=1)
x
```

```python
y = df['Prediction']
y
```

```python
from sklearn.preprocessing import scale
x = scale(x)
```

```python

Q1 = df['Prediction'].quantile(0.25)
Q3 = df['Prediction'].quantile(0.75)
IQR = Q3-Q1
lower_limit=Q1-1.5*IQR
upper_limit=Q3+1.5*IQR
df_without_outliers=df[(df['Prediction']>lower_limit)&(df['Prediction']<upper_limit)]
df_without_outliers
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
# KNN classifier
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=7)

knn.fit(x_train,y_train)
```

```python
y_pred_knn = knn.predict(x_test)
print("PREDICTION :",y_pred_knn)
```

```python
from sklearn import metrics
accuracy_knn = metrics.accuracy_score(y_test,y_pred_knn)
print("accuracy_knn = ",accuracy_knn)
```

```python
# SVM Classifier
from sklearn.svm import SVC
model = SVC(C = 1)
model.fit(x_train,y_train)
```

```python
y_pred_svm = model.predict(x_test)
print("PREDICTION :",y_pred_svm)
```

```python
from sklearn import metrics
accuracy_svm = metrics.accuracy_score(y_test,y_pred_svm)
print("accuracy_svm = ",accuracy_svm)
```
